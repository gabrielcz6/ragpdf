REVISANDO COMO SE VEN LAS IMAGENES

```python
from PIL import Image
import io
import base64



# Función para reducir la resolución de una imagen a 500x500 píxeles en formato JPEG
def resize_image_base64(base64_str, width=500, height=500):
    # Decodificar la imagen base64
    img_data = base64.b64decode(base64_str)
    img = Image.open(io.BytesIO(img_data))
    
    # Redimensionar la imagen a 500x500 píxeles
    img = img.resize((width, height), Image.Resampling.LANCZOS)  # Cambio aquí
    
    # Convertir la imagen a base64 en formato JPEG
    buffered = io.BytesIO()
    img.save(buffered, format="JPEG")  # Cambio a JPEG
    new_base64_str = base64.b64encode(buffered.getvalue()).decode('utf-8')
    
    return new_base64_str

# Función para obtener las imágenes con resolución reducida a 500x500 y formato JPEG
def get_images_base64_resized(chunks, width=500, height=500):
    images_b64_resized = []
    for chunk in chunks:
        if "CompositeElement" in str(type(chunk)):
            chunk_els = chunk.metadata.orig_elements
            for el in chunk_els:
                if "Image" in str(type(el)):
                    base64_str = el.metadata.image_base64
                    resized_image = resize_image_base64(base64_str, width, height)
                    images_b64_resized.append(resized_image)
    return images_b64_resized

# Obtener las imágenes con la resolución de 500x500 y formato JPEG
images = get_images_base64_resized(chunks)


```

```python
import base64

def image_to_base64(image_path):
    with open(image_path, "rb") as image_file:
        # Read and encode the image file to base64
        encoded_image = base64.b64encode(image_file.read())
        # Decode the base64 bytes to a string and return it
        return encoded_image.decode('utf-8')
```

RAG PIPELINE

```python
#2 ejecutables, passthrough y lambda
from langchain_core.runnables import RunnablePassthrough, RunnableLambda
from langchain_core.messages import SystemMessage, HumanMessage
from langchain_openai import ChatOpenAI
from base64 import b64decode
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import ChatPromptTemplate


def parse_docs(docs):
    """Split base64-encoded images and texts"""
    b64 = []
    text = []
    for doc in docs:
        try:
            b64decode(doc)
            b64.append(doc)
        except Exception as e:
            text.append(doc)
    return {"images": b64, "texts": text}

def build_prompt(kwargs):

    docs_by_type = kwargs["context"]
    user_question = kwargs["question"]

    context_text = ""
    if len(docs_by_type["texts"]) > 0:
        for text_element in docs_by_type["texts"]:
            context_text += text_element.text

    # construct prompt with context (including images)
    prompt_template = f"""
    
    Responda a la pregunta :
    
    - basándose únicamente en el siguiente Contexto, 
    - que puede incluir texto, tablas y la siguiente imagen.
    - no alucines, da respuesta solo del Contexto.
    
    Contexto: {context_text}
    Question: {user_question}
    
    """

    prompt_content = [{"type": "text", "text": prompt_template}]

    if len(docs_by_type["images"]) > 0:
        for image in docs_by_type["images"]:
            prompt_content.append(
                {
                    "type": "image_url",
                    "image_url": {"url": f"data:image/jpeg;base64,{image}"},
                }
            )

    return ChatPromptTemplate.from_messages(
        [
            HumanMessage(content=prompt_content),
        ]
    )


chain = (
    {
        "context": retriever | RunnableLambda(parse_docs),
        "question": RunnablePassthrough(),
    }
    | RunnableLambda(build_prompt)
    | ChatOpenAI(model="gpt-4o-mini")
    | StrOutputParser()
)

chain_with_sources = {
    "context": retriever | RunnableLambda(parse_docs),
    "question": RunnablePassthrough(),
} | RunnablePassthrough().assign(
    response=(
        RunnableLambda(build_prompt)
        | ChatOpenAI(model="gpt-4o-mini")
        | StrOutputParser()
    )
)
```
